```python
# Imports
from IPython.display import clear_output
from time import sleep

# Board in its initial state
columns = ' 1  2  3  4  5  6  7'

row_6 = ['( )', '( )', '( )', '( )', '( )', '( )', '( )']
row_5 = ['( )', '( )', '( )', '( )', '( )', '( )', '( )']
row_4 = ['( )', '( )', '( )', '( )', '( )', '( )', '( )']
row_3 = ['( )', '( )', '( )', '( )', '( )', '( )', '( )']
row_2 = ['( )', '( )', '( )', '( )', '( )', '( )', '( )']
row_1 = ['( )', '( )', '( )', '( )', '( )', '( )', '( )']

# Board builder variables
row_6_build = row_6[0]+row_6[1]+row_6[2]+row_6[3]+row_6[4]+row_6[5]+row_6[6]
row_5_build = row_5[0]+row_5[1]+row_5[2]+row_5[3]+row_5[4]+row_5[5]+row_5[6]
row_4_build = row_4[0]+row_4[1]+row_4[2]+row_4[3]+row_4[4]+row_4[5]+row_4[6]
row_3_build = row_3[0]+row_3[1]+row_3[2]+row_3[3]+row_3[4]+row_3[5]+row_3[6]
row_2_build = row_2[0]+row_2[1]+row_2[2]+row_2[3]+row_2[4]+row_2[5]+row_2[6]
row_1_build = row_1[0]+row_1[1]+row_1[2]+row_1[3]+row_1[4]+row_1[5]+row_1[6]

# Starting player
player = 1

# Game status
game_over = False
```

```python
def refresh_board():
    # Update board values
    row_6_build = row_6[0]+row_6[1]+row_6[2]+row_6[3]+row_6[4]+row_6[5]+row_6[6]
    row_5_build = row_5[0]+row_5[1]+row_5[2]+row_5[3]+row_5[4]+row_5[5]+row_5[6]
    row_4_build = row_4[0]+row_4[1]+row_4[2]+row_4[3]+row_4[4]+row_4[5]+row_4[6]
    row_3_build = row_3[0]+row_3[1]+row_3[2]+row_3[3]+row_3[4]+row_3[5]+row_3[6]
    row_2_build = row_2[0]+row_2[1]+row_2[2]+row_2[3]+row_2[4]+row_2[5]+row_2[6]
    row_1_build = row_1[0]+row_1[1]+row_1[2]+row_1[3]+row_1[4]+row_1[5]+row_1[6]
    
    # Rebuild board with updated values
    print(columns)
    print(row_6_build)
    print(row_5_build)
    print(row_4_build)
    print(row_3_build)
    print(row_2_build)
    print(row_1_build)
    
def player_turn():
    global player
    if player == 1:
        get_selection('(O)')
        player = 2
    else:
        get_selection('(@)')
        player = 1
```

```python
def get_selection(player_symbol):
    valid = False
    selection = 0

    refresh_board()
    
    selection = input(f'Player {player_symbol}, select a column: ')
    
    while not valid:
        while not selection.isdecimal():
            clear_output()
            refresh_board()
            selection = input(f'Invalid selection. Player {player_symbol}, please input a number from 1-7: ')
            
        if int(selection) not in range(8):
            clear_output()
            refresh_board()
            selection = input(f'Invalid selection. Player {player_symbol}, please input a number from 1-7: ')
            
        else:
            clear_output()
            valid = True
            place_chip(int(selection), player_symbol)
```

```python
def place_chip(selection, player_symbol):
    if row_1[selection-1] == '( )':
        row_1[selection-1] = player_symbol
        clear_output()
        is_win_condition_met(player_symbol)
    else:
        if row_2[selection-1] == '( )':
            row_2[selection-1] = player_symbol
            clear_output()
            is_win_condition_met(player_symbol)
        else:
            if row_3[selection-1] == '( )':
                row_3[selection-1] = player_symbol
                clear_output()
                is_win_condition_met(player_symbol)
            else:
                if row_4[selection-1] == '( )':
                    row_4[selection-1] = player_symbol
                    clear_output()
                    is_win_condition_met(player_symbol)
                else:
                    if row_5[selection-1] == '( )':
                        row_5[selection-1] = player_symbol
                        clear_output()
                        is_win_condition_met(player_symbol)
                    else:
                        if row_6[selection-1] == '( )':
                            row_6[selection-1] = player_symbol
                            clear_output()
                            is_win_condition_met(player_symbol)
                        else:
                            print(f'Column {selection} is full!')
                            get_selection(player_symbol)
```

```python
def is_win_condition_met(player_symbol):
        
    # Dictionaries to verify win condition (any value containing the same symbol 4 times)
    row_check = {1:[row_1[0],row_1[1],row_1[2],row_1[3]],
            2:[row_1[1],row_1[2],row_1[3],row_1[4]],
            3:[row_1[2],row_1[3],row_1[4],row_1[5]],
            4:[row_1[3],row_1[4],row_1[5],row_1[6]],
             
            5:[row_2[0],row_2[1],row_2[2],row_2[3]],
            6:[row_2[1],row_2[2],row_2[3],row_2[4]],
            7:[row_2[2],row_2[3],row_2[4],row_2[5]],
            8:[row_2[3],row_2[4],row_2[5],row_2[6]],
             
            9:[row_3[0],row_3[1],row_3[2],row_3[3]],
            10:[row_3[1],row_3[2],row_3[3],row_3[4]],
            11:[row_3[2],row_3[3],row_3[4],row_3[5]],
            12:[row_3[3],row_3[4],row_3[5],row_3[6]],
             
            13:[row_4[0],row_4[1],row_4[2],row_4[3]],
            14:[row_4[1],row_4[2],row_4[3],row_4[4]],
            15:[row_4[2],row_4[3],row_4[4],row_4[5]],
            16:[row_4[3],row_4[4],row_4[5],row_4[6]],
             
            17:[row_5[0],row_5[1],row_5[2],row_5[3]],
            18:[row_5[1],row_5[2],row_5[3],row_5[4]],
            19:[row_5[2],row_5[3],row_5[4],row_5[5]],
            20:[row_5[3],row_5[4],row_5[5],row_5[6]],
             
            21:[row_6[0],row_6[1],row_6[2],row_6[3]],
            22:[row_6[1],row_6[2],row_6[3],row_6[4]],
            23:[row_6[2],row_6[3],row_6[4],row_6[5]],
            24:[row_6[3],row_6[4],row_6[5],row_6[6]]}

    column_check = {1:[row_1[0],row_2[0],row_3[0],row_4[0]],
               2:[row_2[0],row_3[0],row_4[0],row_5[0]],
               3:[row_3[0],row_4[0],row_5[0],row_6[0]],
               4:[row_1[1],row_2[1],row_3[1],row_4[1]],
               5:[row_2[1],row_3[1],row_4[1],row_5[1]],
               6:[row_3[1],row_4[1],row_5[1],row_6[1]],
               7:[row_1[2],row_2[2],row_3[2],row_4[2]],
               8:[row_2[2],row_3[2],row_4[2],row_5[2]],
               9:[row_3[2],row_4[2],row_5[2],row_6[2]],
               10:[row_1[3],row_2[3],row_3[3],row_4[3]],
               11:[row_2[3],row_3[3],row_4[3],row_5[3]],
               12:[row_3[3],row_4[3],row_5[3],row_6[3]],
               13:[row_1[4],row_2[4],row_3[4],row_4[4]],
               14:[row_2[4],row_3[4],row_4[4],row_5[4]],
               15:[row_3[4],row_4[4],row_5[4],row_6[4]],
               16:[row_1[5],row_2[5],row_3[5],row_4[5]],
               17:[row_2[5],row_3[5],row_4[5],row_5[5]],
               18:[row_3[5],row_4[5],row_5[5],row_6[5]],
               19:[row_1[6],row_2[6],row_3[6],row_4[6]],
               20:[row_2[6],row_3[6],row_4[6],row_5[6]],
               21:[row_3[6],row_4[6],row_5[6],row_6[6]]}

    diag_check = {1:[row_4[0],row_3[1],row_2[2],row_1[3]],
             2:[row_5[0],row_3[1],row_2[2],row_1[3]],
             3:[row_4[1],row_3[2],row_2[3],row_1[4]],
             4:[row_6[0],row_5[1],row_4[2],row_3[3]],
             5:[row_5[1],row_4[2],row_3[3],row_2[4]],
             6:[row_4[2],row_3[3],row_2[4],row_1[5]],
             7:[row_6[1],row_5[2],row_4[3],row_3[4]],
             8:[row_5[2],row_4[3],row_3[4],row_2[5]],
             9:[row_4[3],row_3[4],row_2[5],row_1[6]],
             10:[row_6[2],row_5[3],row_4[4],row_3[5]],
             11:[row_5[3],row_4[4],row_3[5],row_2[6]],
             12:[row_6[3],row_5[4],row_4[5],row_3[6]],
             13:[row_3[0],row_4[1],row_5[2],row_6[3]],
             14:[row_2[0],row_3[1],row_4[2],row_5[3]],
             15:[row_3[1],row_4[2],row_5[3],row_6[4]],
             16:[row_1[0],row_2[1],row_3[2],row_4[3]],
             17:[row_2[1],row_3[2],row_4[3],row_5[4]],
             18:[row_3[2],row_4[3],row_5[4],row_6[5]],
             19:[row_1[1],row_2[2],row_3[3],row_4[4]],
             20:[row_2[2],row_3[3],row_4[4],row_5[5]],
             21:[row_3[3],row_4[4],row_5[5],row_6[6]],
             22:[row_1[2],row_2[3],row_3[4],row_4[5]],
             23:[row_2[3],row_3[4],row_4[5],row_5[6]],
             24:[row_1[3],row_2[4],row_3[5],row_4[6]]}

    global game_over
    
    for key in row_check.keys():
        if row_check[key] == [player_symbol, player_symbol, player_symbol, player_symbol]:
            game_over = True
            break

    for key in column_check.keys():
        if column_check[key] == [player_symbol, player_symbol, player_symbol, player_symbol]:
            game_over = True
            break
                  
    for key in diag_check.keys():
        if diag_check[key] == [player_symbol, player_symbol, player_symbol, player_symbol]:
            game_over = True
            break
```

```python
def connect_all_4():
    while game_over == False:
        player_turn()

    if player == 2:
        print('Player (O) wins!')
        refresh_board()
    else:
        print('Player (@) wins!')
        refresh_board()
```

```python
connect_all_4()
```
